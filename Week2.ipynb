```
#1. -Tampilkan image dalam channel Red-Green !

img_rg = img.copy()
img_rg[:,:,2] = 0
plt.imshow(img_rg)
```

```
# -Tampilkan image dalam channel Green-Blue
img_gb = img
img_gb[:,:,0] = 0
plt.imshow(img_gb)
```

```
#2. Tampilkan image baris ke 10-100, kolom 10-100!
img_k = img[10:100,10:100]
plt.imshow(img_k)
```

```
#3. Tampilkan image baris ke 5-30, semua kolom, channel Green saja!
plt.imshow(img[5:30])
```

```
#4. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!
img_ktk = cv.cvtColor(img,cv.COLOR_BGR2RGB)
cv.rectangle(img_ktk, pt1=(100,200), pt2=(40,100), color=(0,255,255), thickness=10)
cv.rectangle(img_ktk, pt1=(200,100), pt2=(300,200), color=(0,255,0),thickness=10)
cv.rectangle(img_ktk, pt1=(400,200), pt2=(320,140), color=(255,255,255),thickness=10)
cv.rectangle(img_ktk, pt1=(40,50), pt2=(300,20), color=(255,0,0),thickness=10)
cv.rectangle(img_ktk, pt1=(100,250), pt2=(200,300), color=(0,0,255),thickness=10)
plt.imshow(img_ktk)
```

```
#5. Tampilkan image dengan posisi terbalik!
img_t = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),0)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.imshow(img_t)
```
